# Data Cleaning

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import random
import warnings
%matplotlib inline

def seed_everything(seed):
    random.seed(seed)
    np.random.seed(seed)

seed = 42
seed_everything(seed)
warnings.filterwarnings('ignore')
sns.set()
```

```python
col_names = ["Age group", "Gender", "Category", "Marital status", 'Price', 'Brand name',
       'Aesthetics', 'Technological features', 'Functionality', 'Size',
       'Customisable options', 'Sustainability', "Likelihood", 'Wheels', 'Grilles', 'Headlights', 'Side mirrors', 'Bumpers',
       'Steering wheel', 'Centre compartment', 'Door handles', 'Dashboard', "Amount willing to spend", "Personalisation",
            "Amount willing to pay", "Experience"]
df = pd.read_excel("Product_Survey_Results_Modified.xlsx", index_col=0, parse_dates=True, header=None, names=col_names, skiprows=1)
df = df.reset_index(drop=True).dropna()
df.head()
```

```python
# Preparing hash map for label encoding
age_map = {'20-30': 25,
           '31-40': 35,
           '41-50': 45,
           '51-60': 55}

df["Age group"] = df["Age group"].map(age_map) # replace age group with mean age per group
```

```python
# Assuming each range of values in the age group follows a uniform distribution
df["Age"] = df["Age group"].copy()
df["Age"].values[df["Age"] == 25] = np.random.randint(20, 30, size=len(df["Age"].values[df["Age"] == 25]))
df["Age"].values[df["Age"] == 35] = np.random.randint(31, 40, size=len(df["Age"].values[df["Age"] == 35]))
df["Age"].values[df["Age"] == 45] = np.random.randint(41, 50, size=len(df["Age"].values[df["Age"] == 45]))
df["Age"].values[df["Age"] == 55] = np.random.randint(51, 60, size=len(df["Age"].values[df["Age"] == 55]))
```

```python
# Assuming each range of values in the amount customers are willing to spend on car customisation follows a uniform distribution
df["Amount willing to spend"].values[df["Amount willing to spend"] == "100-500"] = np.random.randint(100, 500, size=len(df["Amount willing to spend"].values[df["Amount willing to spend"] == "100-500"]))
df["Amount willing to spend"].values[df["Amount willing to spend"] == "under 500"] = np.random.randint(0, 500, size=len(df["Amount willing to spend"].values[df["Amount willing to spend"] == "under 500"]))
df["Amount willing to spend"].values[df["Amount willing to spend"] == "500-1000"] = np.random.randint(500, 1000, size=len(df["Amount willing to spend"].values[df["Amount willing to spend"] == "500-1000"]))
df["Amount willing to spend"].values[df["Amount willing to spend"] == "over 1000"] = np.random.randint(1000, size=len(df["Amount willing to spend"].values[df["Amount willing to spend"] == "over 1000"]))
```

```python
# Regardless, we retain the original categories through label encoding. We use the mean (where applicable) as the label.
# However, we group the categories "under 500" and "100-500" together.
df["Spending category"] = df["Amount willing to spend"].copy()
df["Spending category"].values[(df["Spending category"] > 0) & (df["Spending category"] <= 500)] = 250
df["Spending category"].values[(df["Spending category"] > 500) & (df["Spending category"] <= 1000)] = 750
df["Spending category"].values[df["Spending category"] > 1000] = 9999
```

```python
# Assuming each range of values in the amount customers are willing to pay for personalisation follows a uniform distribution
df["Amount willing to pay"].values[df["Amount willing to pay"] == "100-500"] = np.random.randint(100, 500, size=len(df["Amount willing to pay"].values[df["Amount willing to pay"] == "100-500"]))
df["Amount willing to pay"].values[df["Amount willing to pay"] == "under 500"] = np.random.randint(0, 500, size=len(df["Amount willing to pay"].values[df["Amount willing to pay"] == "under 500"]))
df["Amount willing to pay"].values[df["Amount willing to pay"] == "500-1000"] = np.random.randint(500, 1000, size=len(df["Amount willing to pay"].values[df["Amount willing to pay"] == "500-1000"]))
df["Amount willing to pay"].values[df["Amount willing to pay"] == "over 1000"] = np.random.randint(1000, size=len(df["Amount willing to pay"].values[df["Amount willing to pay"] == "over 1000"]))
```

```python
# Regardless, we retain the original categories through label encoding. We use the mean (where applicable) as the label.
# However, we group the categories "under 500" and "100-500" together.
df["Paying category"] = df["Amount willing to pay"].copy()
df["Paying category"].values[(df["Paying category"] > 0) & (df["Paying category"] <= 500)] = 250
df["Paying category"].values[(df["Paying category"] > 500) & (df["Paying category"] <= 1000)] = 750
df["Paying category"].values[df["Paying category"] > 1000] = 9999
```

```python
df["Gender"].values[df["Gender"] == 0] = 1
df = df.drop(["Experience"], axis=1)
df.head()
```

###  2c. and 3c. Among respondents who indicated interest in customisation, is there a relationship between their degree of interest and the amount they are willing to spend on customisation? Can we answer the same question for respondents who indicated interest in personalisation?

```python
df["Spending category"] = df["Spending category"].astype("int")
df["Paying category"] = df["Paying category"].astype("int")
```

Traditionally, to measure the association between two nominal (categorical) variables, we use the Cramer's V statistic. The Cramer's V statistic outputs a value between 0 and 1, where a higher value indicates a higher association.

```python
import scipy.stats as ss

def cramers_corrected_stat(x,y):
    """ calculate Cramers V statistic for categorial-categorial association.
     uses correction from Bergsma and Wicher, 
     Journal of the Korean Statistical Society 42 (2013): 323-328
    """
    result=-1
    if len(x.value_counts())==1 :
        print("First variable is constant")
    elif len(y.value_counts())==1:
        print("Second variable is constant")
    else:   
        conf_matrix=pd.crosstab(x, y)

    if conf_matrix.shape[0]==2:
        correct=False
    else:
        correct=True

    chi2 = ss.chi2_contingency(conf_matrix, correction=correct)[0]

    n = sum(conf_matrix.sum())
    phi2 = chi2/n
    r,k = conf_matrix.shape
    phi2corr = max(0, phi2 - ((k-1)*(r-1))/(n-1))    
    rcorr = r - ((r-1)**2)/(n-1)
    kcorr = k - ((k-1)**2)/(n-1)
    result=np.sqrt(phi2corr / min( (kcorr-1), (rcorr-1)))
    return round(result,6)
```

```python
cvs_customise = cramers_corrected_stat(df["Likelihood"], df["Spending category"])
cvs_personalise = cramers_corrected_stat(df["Personalisation"], df["Paying category"])

print(f"Measure of association between likelihood of customisation and amount willing to spend on customisation is {cvs_customise:.2f}.")
print(f"Measure of association between interest in personalised design and amount willing to pay for personalisation is {cvs_personalise:.2f}.")
```

From this, we can see that the measure of association for both cases are very low (in fact, no association between likelihood of customisation and amount willing to spend on customisation).

Additionally, we make a reasonable assumption that the higher the value of our label, the higher the degree of interest. i.e. We labelled "Likely to customise" as 1, and "Very likely to customise" as 2. 

For degree of interest expressed in designing own components to personalise their cars, we assume that people who only want to do it with professional help have a lower degree of interest than people who readily agree with or without professional help.

With this assumption, we can also apply Pearson's correlation between the variables to see if there is a linear relationship between them.

### Is there a correlation between people who are willing to spend on customisation and people who are willing to pay for personalisation?

```python
custom_corr = df[["Likelihood", "Spending category"]].corr()
personal_corr = df[["Personalisation", "Paying category"]].corr()
total_corr = df[["Likelihood", "Spending category", "Personalisation", "Paying category"]].corr()
```

However, there is a strong correlation between amount willing to spend on customisation and amount willing to pay for personalisation. This means that people who were willing to spend more on customisation were also willing to pay more for personalisation.

# Predictive modelling
Can we find out how much a customer is willing to pay from features such as age group, gender, marital status and car ownership? In the following exercise, we aim to predict how much a customer is willing to spend on customisation and personalised design. We will be using the random forest algorithm here as it is a robust, high-performing predictive algorithm for categorical variables.

```python
new_df = df[['Age group', 'Gender', 'Category', 'Marital status',
                'Spending category', 'Paying category']]
```

```python
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
clf = DecisionTreeClassifier()
rf = RandomForestClassifier()
rf_prime = RandomForestClassifier()

X = new_df.iloc[:, :-2]
y = new_df.iloc[:, -1]
y_prime = new_df.iloc[:, -2]

X_train, X_test, y_train, y_test = train_test_split(X, y, stratify=y, random_state=42)
X_trainp, X_testp, y_trainp, y_testp = train_test_split(X, y_prime, stratify=y_prime, random_state=42)

rf.fit(X_train, y_train)
rf_prime.fit(X_trainp, y_trainp);
```

```python
from sklearn.metrics import classification_report
y_pred = rf.predict(X_test)
y_predp = rf_prime.predict(X_testp)
```

```python
print("Classification report for predicting how much a customer will pay for personalised design")
print(classification_report(y_test, y_pred))
```

```python
print("Classification report for predicting how much a customer will spend on customisation")
print(classification_report(y_testp, y_predp))
```

We can see that on average, our algorithm is right 64% of the time.

# Association Rule Learning
Customers who chose to customise wheels (or any other part) also chose to customise what other components?

```python
df = pd.read_excel("Product_Survey_Results.xlsx")
df.head()
```

```python
asl = df[['Which of the following exterior components would you choose to customise (texture, layout, size, etc)? ',
   'Which of the following interior components would you choose to customise (texture, layout, size, etc)? ']]
asl = asl.rename(columns={'Which of the following exterior components would you choose to customise (texture, layout, size, etc)? ': "Exterior components",
                   'Which of the following interior components would you choose to customise (texture, layout, size, etc)? ': "Interior components"})
asl.head()
```

```python
asl_ext = asl["Exterior components"].str.split(";", expand=True)
asl_int = asl["Interior components"].str.split(";", expand=True)
asl = asl_ext.merge(asl_int, left_index=True, right_index=True)
asl.head()
```

```python
def squeeze_nan(x):
    original_columns = x.index.tolist()

    squeezed = x.dropna()
    squeezed.index = [original_columns[n] for n in range(squeezed.count())]

    return squeezed.reindex(original_columns, fill_value=np.nan)
asl = asl.apply(squeeze_nan, axis=1)
```

```python
components = []
for i in range(0, len(asl)):
    components.append([str(asl.values[i,j]) for j in range(0, len(asl.columns))])
```

```python
# min_support = minimum number rows containing x (any component) / total number of rows
# min_confidence = minimum number rows containing x (any component) and y (complementary component) / total number of rows
# min_lift = min_confidence / min_support
# min_length = minimum number of components in each associated rule
from apyori import apriori
rules = apriori(components, min_support=0.05, min_confidence=0.8, min_lift=3, min_length=2)
```

We will be using the Apriori algorithm for association rule learning between components to find out what other parts do customers, who want to customise "x" component, also want to customise?

From this, we are saying we want to find associated rules between any 2 or more components that are present in at least 5 of the 100 rows (of feedback) with a confidence of 80% (the rule holds true at least 80% of the time).

```python
results = list(rules)
results
```

```python
def inspect(results):
    lhs         = [tuple(result[2][0][0])[0] for result in results]
    rhs         = [tuple(result[2][0][1])[0] for result in results]
    supports    = [result[1] for result in results]
    confidences = [result[2][0][2] for result in results]
    lifts       = [result[2][0][3] for result in results]
    return list(zip(lhs, rhs, supports, confidences, lifts))

results_df = pd.DataFrame(inspect(results), columns = ['Left Hand Side', 'Right Hand Side', 'Support', 'Confidence', 'Lift'])
```

```python
results_df = results_df.replace('nan', np.nan).dropna()
results_df.sort_values('Support', ascending=False).drop_duplicates(subset=['Left Hand Side', 'Right Hand Side'])
```

According to the top 3 rules, we can see that 
- Customers who want to customise side mirrors also want to customise steering wheels 100% of the time, and this rule appears in 7 feedbacks. 
- Customers who want to customise side mirrors also want to customise door handles 87.5% of the time, and this rule appears in 7 feedbacks.
- Customers who want to customise headlights also want to customise steering wheels 87.5% of the time, and this rule appears in 7 feedbacks.
